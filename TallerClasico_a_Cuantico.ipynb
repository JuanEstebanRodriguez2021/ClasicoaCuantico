```python
import random
import matplotlib.pyplot as plt

def double_slit_experiment(num_trials):
    

    targets = []
    for _ in range(num_trials):
        slit = random.randint(0, 1)

        if slit == 0:
            target = random.choice([0, 2, 7])
        else:
            target = random.choice([1, 3, 4, 5, 6])

        targets.append(target)
    return targets

def visualize_results(targets):
    
    counts = [0] * 8
    for target in targets:
        counts[target] += 1

    plt.bar(range(8), counts)
    plt.xlabel("Target")
    plt.ylabel("Intensity")
    plt.title("Double-Slit Experiment Results")
    plt.show()


num_trials = 10000
targets = double_slit_experiment(num_trials)


visualize_results(targets)
```

```python
import numpy as np
import matplotlib.pyplot as plt

states = [1/np.sqrt(2), 1/np.sqrt(2)]

paths = [
    [1, 3, 7],
    [1, 4, 7],
    [1, 5, 7],
    [2, 3, 7],
    [2, 4, 7],
    [2, 5, 7],
    [2, 6, 7]
]
amplitudes = [
    (1+1j)/np.sqrt(6), (1-1j)/np.sqrt(6), -1/np.sqrt(6),
    (1+1j)/np.sqrt(6), (1-1j)/np.sqrt(6), -1/np.sqrt(6),
    (1-1j)/np.sqrt(6)
]

probabilities = []
for path in paths:
    amplitude = states[0]
    for i, step in enumerate(path):
        amplitude *= amplitudes[i]
    probability = np.abs(amplitude)**2
    probabilities.append(probability)

plt.bar(range(1, 8), probabilities)
plt.xlabel("Target")
plt.ylabel("Probability")
plt.title("Double Slit Experiment")
plt.show()
```

3. Simulación del Experimento de la doble rendija

```python
import numpy as np
import matplotlib.pyplot as plt

slit_distance = 0.5e-3
wavelength = 650e-9
screen_distance = 1.0
screen_width = 0.02
num_points = 1000

screen_points = np.linspace(-screen_width / 2, screen_width / 2, num_points)
intensity = np.zeros(num_points)
slit_positions = [-slit_distance / 2, slit_distance / 2]

for slit in slit_positions:
    distances = np.sqrt((screen_distance)**2 + (screen_points - slit)**2)
    phases = (2 * np.pi / wavelength) * distances
    intensity += np.cos(phases)

intensity = intensity**2

plt.figure(figsize=(10, 6))
plt.plot(screen_points * 1e3, intensity, label='Interference Pattern')
plt.xlabel('Position on Screen (mm)')
plt.ylabel('Intensity (arbitrary units)')
plt.title('Double Slit Experiment Simulation: Wave Interference Pattern')
plt.legend()
plt.grid(True)
plt.show()
```

4. Informe de Experimento de Doble Rendija


Objetivo:

El objetivo del experimento de la doble rendija es demostrar el principio de interferencia de ondas, confirmando la naturaleza ondulatoria de la luz, y explorar cómo las ondas interactúan para crear patrones de interferencia. Se pretende medir la distancia entre los máximos de interferencia y analizar los factores que influyen en este patrón.

Fundamento Teórico:

El experimento de la doble rendija fue propuesto por primera vez por Thomas Young en 1801 para demostrar la naturaleza ondulatoria de la luz. Según el modelo ondulatorio, cuando la luz pasa por dos rendijas estrechas y coherentes, las ondas de luz que emergen de cada rendija interfieren entre sí. Esta interferencia puede ser constructiva o destructiva, dependiendo de las diferencias en el camino recorrido por las ondas desde las rendijas hasta la pantalla de observación.

La ecuación principal que rige el patrón de interferencia es:

Δ𝐿 = 𝑑 * sin(θ)
⁡Donde:

Δ𝐿 es la diferencia en la longitud de los caminos ópticos.
𝑑 es la distancia entre las dos rendijas.
𝜃 es el ángulo de interferencia observado.
Para condiciones de interferencia constructiva (máximos), la diferencia de camino óptico es igual a un múltiplo entero de la longitud de onda (𝜆):

𝑑 * sin(𝜃) = 𝑚𝜆

Donde 
𝑚
m es un número entero (orden de la interferencia).

Para la interferencia destructiva (mínimos), la diferencia de camino es un múltiplo semientero de la longitud de onda:

𝑑 * sin(𝜃) = (𝑚 + 1/2)λ

- Alineación inexacta del láser con las rendijas.
